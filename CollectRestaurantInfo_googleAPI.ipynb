```python
#Install google maps package 
#!pip install --use-feature=2020-resolver --target=/usrs/c/anaconda3/lib/site-packages --upgrade googlemaps
import googlemaps
from pprint import pprint
import pandas as pd
from time import sleep
import csv
import json
```

```python
API_KEY = '**************' 

gmaps = googlemaps.Client(key=API_KEY)
loc = [] 
ids = []
stores_info = []

#I downloaded a file which lists all roads in Taiwan beforehand
#And, read the file here
with open('109臺東road.csv') as csvfile:
    rows = csv.reader(csvfile)
    roadlist = []

    for row in rows:
        roadlist.append(row[0].replace('\t',''))


#Recognize the longitude and latitude of all roads    
for road in roadlist:
                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                
    geocode_result = gmaps.geocode(road)
    loc = geocode_result[0]['geometry']['location']
   
    #Use the longitude and latitude of the center of all roads to collect restaurants' URLs within 5 meters radius
    for place in gmaps.places_nearby(keyword='restaurant', location= loc, radius= 500)['results']:
        ids.append(place['place_id'])

#Use "set" to avoid duplicated ids
ids = list(set(ids))


#Use id to extract restaurants' information
for id in ids:
     stores_info.append(gmaps.place(place_id=id,language= 'zh-TW')['result'])

oput = pd.DataFrame.from_dict(stores_info)

oput
```

```python
#Save the file in both CSV and JSON version
df = pd.DataFrame(stores_info)
df.to_csv("Taitung.csv", index=False)
with open("Taitung.json", "w",encoding="utf-8") as f:
    json.dump(stores_info, f, ensure_ascii=False)
```
